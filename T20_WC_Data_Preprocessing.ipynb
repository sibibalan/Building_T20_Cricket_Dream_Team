```python
import pandas as pd
import numpy as np
import re
```

```python
# t20_wc_match_results.csv is imported and the date column is converted into datetime object
df_match = pd.read_csv('t20_wc_match_results.csv',parse_dates=['Match Date'])
df_match.head()
```

```python
# rename the Scorecard column to match_id.
df_match.rename({'Scorecard':'match_id'},axis = 1, inplace=True)
df_match.head()
```

```python
# t20_wc_batting_summary.csv is imported
df_batting = pd.read_csv('t20_wc_batting_summary.csv')
df_batting.head()
```

```python
# renamed multiple columns and
# changed the value of dismissal column into 0 for 'not out' and 1 for 'out andstored in a new column named 'out'
# alsodropped unnecessary columns.
df_batting.rename({'4s':'fours',
                   '6s':'sixes'},axis = 1, inplace=True)
df_batting['out'] = df_batting['dismissal'].apply(lambda x: 0 if x.strip() == 'not out' else 1)
df_batting.drop(columns=["dismissal"],inplace=True)
```

```python
# removed special characters in the batsman_name column and also captain notation( (c) )
def remove_special_chr(text,pattern):
  r = re.compile(pattern)
  return r.sub('',text).strip()

pattern1 = r'[^a-zA-Z -() ]+'
pattern2 = r'\(c\)'
pattern3 = r'\(c\)|[^-\w\s]'
pattern4 = r'\(c\)|[^a-zA-Z !@#$%&*_+-=|\:";<>,./()[\]{}\']'

# df_batting['batsman_name'] =  df_batting['batsman_name'].apply(lambda name : remove_special_chr(text=name,pattern=pattern1))
# df_batting['batsman_name'] = df_batting['batsman_name'].apply(lambda name :remove_special_chr(text=name,pattern=pattern2))
df_batting['batsman_name'] =  df_batting['batsman_name'].apply(lambda name : remove_special_chr(text=name,pattern=pattern4))
```

```python
df_batting.head()
```

```python
df_bowling = pd.read_csv('t20_wc_bowling_summary.csv')
df_bowling.head()
```

```python
# renamed various columns.
df_bowling.rename({'0s':'zeros',
                    '4s':'fours',
                   '6s':'sixes'},axis = 1, inplace=True)
```

```python
df_player = pd.read_csv('t20_wc_player_info.csv')
df_player.head()
```

```python
# converted the NaN into empty strings.
df_player['description'].replace(np.nan,'', inplace=True)
```

```python
# store the processed dfs into new csv files inorder to imported by powerbi
df_match.to_csv('dim_match_summary.csv',index=False)
df_batting.to_csv('fact_batting_summary.csv',index=False)
df_bowling.to_csv('fact_bowling_summary.csv',index=False)
df_player.to_csv('dim_players.csv',index=False)
```
